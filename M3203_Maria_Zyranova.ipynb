## Замыкание/размыкание цепи.


```python
import numpy as np
import plotly.graph_objects as go
```

```python
# @title Параметры
# @markdown Индуктивность, L (Гн):
L = 9 # @param {type:"integer"}
# @markdown Сопративление, R (Ом):
R = 5 # @param {type:"integer"}
# @markdown ЭДС источника тока, ε (В):
eps = 5 # @param {type:"integer"}
# @markdown Время, t (c):
t = 10 # @param {type:"integer"}
```

```python
if L <= 0:
    raise Exception("Введите неотрицательную индуктивность")
if t <= 0:
    raise Exception("Введите положительное время")
if R <= 0:
    raise Exception("Введите неотрицательное сопротивление")
if eps < 0:
    raise Exception("Введите неотрицательную ЭДС источника тока")
```

```python
step = t / 1000
x = np.linspace(0, t, 1000)
tau = L / R
i_0 = eps / R

open_y = i_0 * np.exp(-x / tau)
close_y = i_0 * (1 - np.exp(-x / tau))
```

## Сила тока при размыкании цепи

```python
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=x, y=open_y, mode='lines+markers', marker_color='red'))
fig1.update_layout(width=800, height=420, xaxis_title='Время, с', yaxis_title='Сила тока, А')
fig1.show()
```

## Сила тока при замыкании цепи

```python
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=x, y=close_y, mode='lines+markers', marker_color='blue'))
fig2.update_layout(width=800, height=420, xaxis_title='Время, с', yaxis_title='Сила тока, А')
fig2.show()
```
